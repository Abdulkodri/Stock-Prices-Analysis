## Handling Outliers by Replacing with Boundaries

```python
def replace_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return series.apply(lambda x: lower_bound if x < lower_bound else upper_bound if x > upper_bound else x)

# Apply the function to the relevant columns
df['Close_M'] = replace_outliers(df['Close_M'])
df['Close_V'] = replace_outliers(df['Close_V'])
```

## Check normality using Anderson-Darling Test and normalize using PowerTransformer if necessary

```python
from sklearn.preprocessing import PowerTransformer
def check_normality_and_normalize(df, columns):
    pt = PowerTransformer()
    for col in columns:
        result = anderson(df[col])
        print(f'Anderson-Darling Test for {col}: Statistic={result.statistic:.3f}, Critical values={result.critical_values}')
        if result.statistic > result.critical_values[2]:  # Using 5% significance level
            print(f'{col} is not normally distributed. Applying PowerTransformer normalization...')
            df[[col]] = pt.fit_transform(df[[col]].values.reshape(-1, 1))
        else:
            print(f'{col} is normally distributed.')
    return df

# Columns to check and normalize
columns_to_check = ['Close_M', 'Close_V']

# Apply the function to the DataFrame
df = check_normality_and_normalize(df, columns_to_check)
```

```python
# Verify the normalization
print(df.head())
```

## Re-check normality using Anderson-Darling test and plot histograms and Q-Q plots

```python
from scipy.stats import anderson, probplot
def verify_normalization(df, columns):
    for col in columns:
        result = anderson(df[col])
        print(f'Anderson-Darling Test after normalization for {col}: Statistic={result.statistic:.3f}, Critical values={result.critical_values}')
        
        plt.figure(figsize=(12, 5))

        plt.subplot(1, 2, 1)
        sns.histplot(df[col], kde=True)
        plt.title(f'Histogram of {col}')

        plt.subplot(1, 2, 2)
        probplot(df[col], dist="norm", plot=plt)
        plt.title(f'Q-Q Plot of {col}')

        plt.show()

# Columns to check and normalize
columns_to_check = ['Close_M', 'Close_V']

# Apply the function to the DataFrame
df = check_normality_and_normalize(df, columns_to_check)

# Verify the normalization
verify_normalization(df, columns_to_check)
```

```python
# Feature Engineering (Creating Moving Averages)
df['MA_Close_M'] = df['Close_M'].rolling(window=10).mean()
df['MA_Close_V'] = df['Close_V'].rolling(window=10).mean()
df.dropna(inplace=True)
```

```python
# Prepare data for modeling
features = ['Open_M', 'High_M', 'Low_M', 'Volume_M', 'Open_V', 'High_V', 'Low_V', 'Volume_V', 'MA_Close_M', 'MA_Close_V']
target_M = 'Close_M'
target_V = 'Close_V'
```

```python
X = df[features]
y_M = df[target_M]
y_V = df[target_V]
```

```python
# Split data into training and testing sets
X_train, X_test, y_train_M, y_test_M = train_test_split(X, y_M, test_size=0.2, random_state=42)
X_train, X_test, y_train_V, y_test_V = train_test_split(X, y_V, test_size=0.2, random_state=42)
```

```python
# Initialize models
rf = RandomForestRegressor(random_state=42)
lr = LinearRegression()
dt = DecisionTreeRegressor(random_state=42)
```

```python
# Train and evaluate models for Close_M
models = {'Random Forest': rf, 'Linear Regression': lr, 'Decision Tree': dt}
results_M = {}
for model_name, model in models.items():
    model.fit(X_train, y_train_M)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test_M, predictions)
    r2 = r2_score(y_test_M, predictions)
    results_M[model_name] = {'MSE': mse, 'R^2': r2}
```

```python
# Train and evaluate models for Close_V
results_V = {}
for model_name, model in models.items():
    model.fit(X_train, y_train_V)
    predictions = model.predict(X_test)
    mse = mean_squared_error(y_test_V, predictions)
    r2 = r2_score(y_test_V, predictions)
    results_V[model_name] = {'MSE': mse, 'R^2': r2}
```

```python
# Display results
print("Model Performance for Close_M:")
print(pd.DataFrame(results_M))
```

```python
print("\nModel Performance for Close_V:")
print(pd.DataFrame(results_V))
```

```python
# Fit ARIMA Model and Forecast Future Prices for Close_M
arima_model_M = ARIMA(df['Close_M'], order=(5, 1, 0))
arima_result_M = arima_model_M.fit()
```

```python
# Forecasting for 30 days
forecast_M = arima_result_M.forecast(steps=30)
forecast_dates_M = pd.date_range(df.index[-1], periods=30, freq='B')
```

```python
# Forecasting for 30 days
forecast_V = arima_result_V.forecast(steps=30)
forecast_dates_V = pd.date_range(df.index[-1], periods=30, freq='B')
```
